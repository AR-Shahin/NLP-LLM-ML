```python
import logging
from aiogram import Bot, Dispatcher, types
from aiogram.types import Message
from aiogram.filters import Command
from aiogram import F
from dotenv import load_dotenv
import asyncio
import os


load_dotenv()
TELEGRAM_BOT_TOKEN = os.getenv("TELEGRAM_BOT_TOKEN")
```

```python
bot = Bot(token=TELEGRAM_BOT_TOKEN)
# Initialize dispatcher
dp = Dispatcher()
# Configure logging
logging.basicConfig(level=logging.INFO)

# Handler for the /start and /help commands
@dp.message(Command(commands=['start', 'help','ars']))
async def command_start_handler(message: Message):
    """
    This handler receives messages with `/start` or `/help` command
    """
    await message.answer("Hello\nI am Echo Bot!\nPowered by Shahin.")

# Handler for echoing any text message
@dp.message(F.text)
async def echo(message: Message):
    """
    This will return echo of the received message
    """
    await message.answer(message.text)

# Main function to start polling
async def main():
    await dp.start_polling(bot)

# Detect if we're in an environment with an active event loop (like Jupyter)
# if __name__ == "__main__":
#     try:
#         asyncio.run(main())  # For standard environments
#     except RuntimeError:  # If already in a running event loop
#         loop = asyncio.get_event_loop()
#         loop.create_task(main())
```
